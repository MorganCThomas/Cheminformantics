# Structure-aware generative molecular design: from 3D or 2D?

This notebook contains the code to conduct the analysis seen in the blog post (https://cheminformantics.blogspot.com/2024/12/structure-aware-generative-molecular.html) including:
- Geometric metrics
- Chemical metrics
- A little peek at molecules in 2D

#### Imports and notebook functions

```python
import re
import json
import random
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns ; sns.set_context("notebook")
from tqdm import tqdm
from rdkit.Chem import AllChem as Chem
from rdkit.Chem import Draw, rdDepictor
rdDepictor.SetPreferCoordGen(True)
from IPython.display import display

from moleval.utils import read_smiles
from moleval.metrics.metrics import GetMetrics
from moleval.metrics.posecheck import PoseCheck
from moleval.metrics.metrics_utils import mapper
```

```python
# Load
def load_mols(file_path):
    with Chem.SDMolSupplier(file_path) as suppl:
        print(f"Loading {len(suppl)} mols")
        mols = [mol for mol in suppl if mol]
    print(f"Successfully loaded {len(mols)} mols")
    return mols

def show_atom_number(mol, label='atomLabel'):
    for atom in mol.GetAtoms():
        atom.SetProp(label, str(atom.GetIdx()))
    return mol

def fix_radicals_to_hs(mol):
    for atom in mol.GetAtoms():
        if atom.GetNumRadicalElectrons() > 0:
            atom.SetNumExplicitHs(atom.GetNumRadicalElectrons())
            atom.SetNumRadicalElectrons(0)
    Chem.SanitizeMol(mol)
    return mol
```

## Calculate metrics

Calculate all geometry and chemistry metrics using the de novo molecules generated

#### Run PoseCheck

```python
names = {
    "AHC": "data/AHC2_Top1000.sdf",
    "AHC_rDock": "data/AHC_rDock.sdf",
    "Autogrow4": "data/Autogrow4_Top1000.sdf",
    "3D-MCTS": "data/3D-MCTS.sdf",
    "Lingo3DMol": "data/Lingo3DMol.sdf",
    "Pocket2Mol": "data/Pocket2Mol.sdf",
    "TargetDiff": "data/TargetDiff.sdf",
}

for name, fpath in tqdm(names.items()):
    result = {"model": name}

    pc = PoseCheck(n_jobs=12)
    pc.load_protein_from_pdb('data/6CM4/6cm4p_rec.pdb', reduce=False)

    mols = load_mols(fpath)
    # FIX mols with radicals #
    mols = [fix_radicals_to_hs(m) for m in mols]
    pc.load_ligands_from_mols(mols, add_hs=True)
    
    result.update(
        {
            "Clashes": pc.calculate_clashes(),
            "StrainEnergy": pc.calculate_strain_energy(),
            "TanimotoIFPSimilarityPC": pc.calculate_interaction_similarity(
                ref_lig_path = "data/6CM4/6cm4p_lig.sdf",
                similarity_func = "Tanimoto",
                count = False
            )
        }
    )
    with open(f"data/{name}.posecheck", "wt") as f:
        json.dump(result, f, indent=2)
```

#### Run MolEval metrics

```python
# Load ref smiles
D2_ref = read_smiles("data/6CM4/001_001_004_002-drd2_human-5098_LIG")
ref = read_smiles("data/ChEMBL28p_curation.smi.gz")

names = {
    "AHC": "data/AHC_Top1000.sdf",
    "AHC_rDock": "data/AHC_rDock.sdf",
    "Autogrow4": "data/Autogrow4_Top1000.sdf",
    "3D-MCTS": "data/3D-MCTS.sdf",
    "Lingo3DMol": "data/Lingo3DMol.sdf",
    "Pocket2Mol": "data/Pocket2Mol.sdf",
    "TargetDiff": "data/TargetDiff.sdf",
}

mengine = GetMetrics(
    test=ref,
    target=D2_ref,
    run_fcd=False,
    n_jobs=10
)

for name, fpath in tqdm(names.items()):
    mols = load_mols(fpath)
    mols = [fix_radicals_to_hs(mol) for mol in mols]
    result = mengine.calculate(
            gen=mols, calc_valid=True, calc_unique=True, return_stats=True
        )
    with open(f'data/{name}.moleval', 'wt') as f:
        json.dump(result, f, indent=2)
```

```python
def bulk_manhatten_dist(x, y, norm=False):
    """
    If binary, the same as hamming distance.
    Let's also normalize by length of x (should be the same as y's)
    """
    x = np.array(x)
    ys = [np.array(_y) for _y in y]
    N = len(x)
    assert all([len(_) == N for _ in y])
    sims = []
    for y in ys:
        dist = np.sum(np.abs(y-x))
        if norm:
            dist = 1 - (dist / N)
        sims.append(dist)
    return sims
```
